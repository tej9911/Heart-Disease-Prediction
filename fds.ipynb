```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=2)
```

```python
# Check the unique values in data_output
unique_classes = np.unique(data_output)
print("Unique classes in data_output:", unique_classes)

if len(unique_classes) <= 1:
    print("Not enough classes for SMOTE. Please ensure your dataset includes both classes.")
else:
    # Proceed with SMOTE
    data_input, data_output = sm.fit_resample(data_input, data_output.ravel())
```
